```python
fuzzy_matches_rw_esewa = rw_df.crossJoin(esewa_df).withColumn(
    "Distance",
    levenshtein(col("rw_Identifier"), col("esewa_Identifier"))
).filter(col("Distance") <= 1)
```

```python
def extract_destination_ac(transaction_detail):
    if 'LOAD' in transaction_detail:
        match = re.search(r'LOAD/(\d+),', transaction_detail)
        if match:
            return match.group(1)
    return None

# Function to extract the destination name
def extract_destination_name(transaction_detail):
    if 'LOAD' in transaction_detail:
        parts = transaction_detail.split(' ')
        if len(parts) > 1:
            return parts[0]
    return None
```

```python
b['targetACC'] = b['TRANSACTION_DETAILS'].apply(extract_destination_ac)
b['targetProvider'] = b['TRANSACTION_DETAILS'].apply(extract_destination_name)
```
