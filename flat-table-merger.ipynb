```python
def extract_and_format_date(date_str):
    if date_str.startswith('java.util.GregorianCalendar'):
        pattern = r'YEAR=(\d+),MONTH=(\d+),.*?DAY_OF_MONTH=(\d+)'
        match = re.search(pattern, date_str)
        if match:
            year, month, day = match.groups()
            month = int(month) + 1  # Months are 0-based in GregorianCalendar
            return datetime(year=int(year), month=month, day=int(day)).strftime('%Y-%m-%d')
    else:
        try:
            # Assuming the date format is dd/mm/yyyy
            return pd.to_datetime(date_str, format='%d/%m/%Y').strftime('%Y-%m-%d')
        except ValueError:
            # If the format is not correct, return the original string
            return date_str

# Apply the function to the DataFrame
eb['Date'] = eb['Date'].apply(extract_and_format_date)
```
