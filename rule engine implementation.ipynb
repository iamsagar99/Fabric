```python

import pandas as pd
import numpy as np
from datetime import timedelta

def flag_transactions(data):
    data['Flag'] = 'False'
    data['Identifier'] = None

    data['Date'] = pd.to_datetime(data['Date'])
    
    data['Month'] = data['Date'].dt.month
    data['Day'] = data['Date'].dt.day

    identifier = 0

    grouped = data.groupby('CustomerId')

    for customer_id, group in grouped:
        group = group.sort_values(by='Date')
        
        amount_day_grouped = group.groupby(['Amount', 'Day'])
        
        for (amount, day), amount_group in amount_day_grouped:
            month_counts = amount_group['Month'].value_counts()
            repeating_months = month_counts[month_counts > 2]
            
            if len(repeating_months) > 0:
                baseline_transaction = amount_group.iloc[0]
                data.loc[baseline_transaction.name, 'Flag'] = 'True'
                data.loc[baseline_transaction.name, 'Identifier'] = identifier
                
                baseline_amount = baseline_transaction['Amount']
                baseline_day = baseline_transaction['Day']
                baseline_date = baseline_transaction['Date']
                
                for month in repeating_months.index:
                    if month != baseline_transaction['Month']:
                        month_group = group[group['Month'] == month]

                        potential_regulars = pd.DataFrame()

                        # Collect transactions within ±5 days and ±10% amount range
                        for i in range(-5, 6):
                            if baseline_day + i < 1:
                                prev_month_days = (baseline_date - pd.DateOffset(months=1)).days_in_month
                                check_day = prev_month_days + (baseline_day + i)
                                check_month = baseline_date.month - 1
                            elif baseline_day + i > 30:
                                check_day = (baseline_day + i) % 30
                                check_month = baseline_date.month + 1
                            else:
                                check_day = baseline_day + i
                                check_month = baseline_date.month

                            check_amount_min = baseline_amount * 0.9
                            check_amount_max = baseline_amount * 1.1

                            potential_regulars = pd.concat([potential_regulars, month_group[
                                (month_group['Day'] == check_day) & 
                                (month_group['Amount'] >= check_amount_min) & 
                                (month_group['Amount'] <= check_amount_max)
                            ]])

                        if not potential_regulars.empty:
                            # Calculate the distance `d`
                            potential_regulars['d'] = np.sqrt(
                                (baseline_amount - potential_regulars['Amount'])**2 +
                                (baseline_day - potential_regulars['Day'])**2
                            )
                            
                            # Select the transaction with the smallest distance
                            regular_income_transaction = potential_regulars.loc[potential_regulars['d'].idxmin()]
                            data.loc[regular_income_transaction.name, 'Flag'] = 'True'
                            data.loc[regular_income_transaction.name, 'Identifier'] = f"{identifier}-g"
                
                identifier += 1

    return data
```

```python
flagged_data = flag_transactions(data)


```
